# Handle Catogorical Variables 

```python
data["ocean_proximity"].value_counts()
```

```python
pd.get_dummies(data["ocean_proximity"]).head()
```

```python
dummy = pd.get_dummies(data["ocean_proximity"])
```

```python
dummy.drop("ISLAND",axis=1,inplace=True)
```

```python
dummy.head()
```

```python
data["<1H OCEAN"] = dummy["<1H OCEAN"].values
data["INLAND"] = dummy["INLAND"].values
data["NEAR BAY"] = dummy["NEAR BAY"].values
data["NEAR "] = dummy["NEAR OCEAN"].values
```

```python
data.head()
```

```python
train_ft = data.drop(["median_house_value","ocean_proximity"],axis=1)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
imputer.fit(train_ft)
```

```python
imputer.statistics_
```

```python
train_ft.median().values
```

```python
x = imputer.transform(train_ft)
```

```python
train_new_set = pd.DataFrame(x,columns=train_ft.columns)
train_new_set.head()
```

```python
train_new_set.isna().sum()
```

```python
train_new_set.info()
```

```python
data.shape
```

```python
train_new_set.shape
```

```python
x = train_new_set.values
y = data["median_house_value"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

# Split the Data

```python
x_train.shape
```

```python
y_test.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lrig = LinearRegression ()
lrig.fit(x_train,y_train)
```

```python
predictions = lrig.predict(x_test[:10])
predictions 
```

```python
y_train[:10]
```

```python
data = {"predicted": predictions,"actual":y_test[:10].values,"diff":(predictions-y_train[:10])}
error_data = pd.DataFrame(data=data)
error_data
```

# Model Evaluation 

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
pred = lrig.predict(x_test)
```

```python
mse = mean_squared_error(y_test,pred)
np.sqrt(mse)
```

```python
mae = mean_absolute_error(y_test,pred)
mae
```

```python
r2 = r2_score(y_test,pred)
r2
```

```python
score = cross_val_score(lrig,x_train,y_train, scoring="neg_mean_squared_error",cv=7)
rmse_score = np.sqrt(-score)
```
